# 5 - Scenario 1
In this scenario, we need to determine the clients we should contact in our next campaign (who are they and their total number) in order to maximize the term deposit subscriptions while avoiding annoyance by carefully managing the frequency of contacts and considering those categories.  


**Objective:** Maximize the number of term deposit subscriptions

**Decision to be made:** The decision-maker must decide the frequency of contacting clients to avoid annoyance


## 5.1 - Mathematical model

#### **Objective Function:**

\begin{align*}
\text{Maximize: } & \sum_{i=1}^{n} Z_i \cdot P_i
\end{align*}


#### **Constraints:**

**Budget constraint:** The total costs of contacts should not exceed the total budget
$$
\sum_{i=1}^{n} (D_i \cdot PC \cdot Z_{i}) \leq TB
$$

**Contact frequency constraint:** Each client can only be contacted if targeted

$$
Z_i  \leq X_{i} \text { for all } i
$$

**Previous number of contacts:** Ensures that clients are contacted only if the previous contacts do net exceed 3

$$
previous_{i} * Z_{i} \leq {\text{3 }} {\text {for all i}}
$$

**Employed category constraint:** It limits the maximum number of employee clients to be targeted to prevent overwhelming a specific group
$$
\sum_{i=1}^{n} X_i \leq  MaxEmp,  {\text{ for all i if category[i] ="employed"}}
$$

**Student category constraint:** It sets a minimum on the number of student clients to be targeted to encourage term deposit subscriptions amongst them

$$
\sum_{i=1}^{n} X_i \geq MinStud,  {\text{ for all i if category[i] ="student"}}
$$

**Retired category constraint:** It sets a minimum on the number of retired clients to be targeted to encourage term deposit subscriptions amongst them

$$
\sum_{i=1}^{n} X_i \geq MinRet,  {\text{ for all i if category[i] ="retired"}}
$$

**Unemployed category constraint:**  It limits the maximum number of employee clients to be targeted to prevent overwhelming a specific group

$$
\sum_{i=1}^{n} X_i \leq MaxUnemp,  {\text{ for all i if category[i] ="unemployed"}}
$$


**Binary variable constraint:**
Each decision variable should be binary, taking values either 0 or 1.
$$
X_{i} \in \{0, 1\}, \text{ for all } i
$$

$$
Z_{i} \in \{0, 1\}, \text{ for all } i
$$



### 5.1.1 - Preliminary operations


```python
!pip install pulp
import pandas as pd
import pulp
from pulp import *
```

### 5.1.2 - Sets


```python
df=pd.read_csv("/content/MDMM - Dataset SC1.csv")
n = len(df)
previous = df['previous'].tolist()
category = df['job'].tolist()
client_ids = df['client_id'].tolist()
```

### 5.1.3 - Parameters
\begin{align*}
& Probabilities_i \in [0, 1]: \text{ expected yield for targeting client } i \\
& D_i: \text{ Duration of call per client} \\
& PC: \text{ Phone Call Cost (per minute) = 0.16 EUR/minute} \\
& TB: \text{ Total budget for the campaign} \\
& MaxEmp : \text{ Maximum number of employee clients to be targeted} \\
& MinStud : \text{ Minimum number of student clients to be targeted} \\
& MaxUnemp : \text{ Maximum number of unemployed clients to be targeted} \\
& MinRet : \text{ Minimum number of retired clients to be targeted} \\
\end{align*}



```python
D = df['duration'].tolist()
probabilities = df['Probability of subscribing to the term deposit'].tolist()
PC = 0.16
TB = 50000
MaxEmp = 5450
MinStud = 130
MaxUnemp = 120
MinRet = 300
```

### 5.1.4 - Variables

\begin{align*}
& X_i \in \{0, 1\}: \text{ 1 if client } i \text{ is targeted in this campaign, 0 otherwise.} \\
& Z_i \in \{0, 1\}: \text{ 1 if client } i \text{ is contacted in this campaign, 0 otherwise.}
\end{align*}


```python
# Initialize the problem
prob = LpProblem("ClientContactOptimization", LpMaximize)

X = LpVariable.dicts("Targeted", range(n), 0, 1, LpBinary)
Z = LpVariable.dicts("Contacted", range(n), 0, 1, LpBinary)
```

### 5.1.5 - Constraints and Objective(s)


```python
probabilities = [float(val.replace(',', '.')) for val in probabilities]

# Objective function
prob += lpSum([probabilities[i] * Z[i] for i in range(n)])

# Constraints for selecting subsets from each category
emp_indices = [i for i in range(n) if category[i] == "employed"]
prob += lpSum(X[i] for i in emp_indices) <= MaxEmp

stud_indices = [i for i in range(n) if category[i] == "student"]
prob += lpSum(X[i] for i in stud_indices) >= MinStud

unemp_indices = [i for i in range(n) if category[i] == "unemployed"]
prob += lpSum(X[i] for i in unemp_indices) <= MaxUnemp

ret_indices = [i for i in range(n) if category[i] == "retired"]
prob += lpSum(X[i] for i in ret_indices) >= MinRet

# Ensure that if a person must be contacted only if they are targeted
for i in range(n):
    prob += Z[i] <= X[i]

# Previous number of contacts constraint
for i in range(n):
    prob += previous[i] * Z[i] <= 3

# Budget constraint
prob += lpSum([D[i] * PC * Z[i] for i in range(n)]) <= TB
```

## 5.2 - Model solution


```python
# Solve the optimization problem
prob.solve()

import csv

# Create a list to hold the results
results = []

# Print the results
print("Status:", LpStatus[prob.status])
print("Optimal solution:")
total_people_targeted = 0
total_people_contacted = 0
for v in prob.variables():
    if v.name.startswith("Targeted") and v.varValue == 1:
        total_people_targeted += 1
    elif v.name.startswith("Contacted") and v.varValue == 1:
        total_people_contacted += 1
        client_id = client_ids[int(v.name.split("_")[1])]  # Extract client ID from variable name
        print("Client_",client_id, "=", v.varValue)
        results.append(("To be Contacted", client_id))

# Write the results to a CSV file
with open("results.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Action", "Client ID"])  # Write the header
    writer.writerows(results)  # Write the results
```

```python
print("Total people targeted:", total_people_targeted)
print("Total people contacted:", total_people_contacted)
print("Value of the objective function:", value(prob.objective)) #This value represents the maximum sum of probabilities of subscribing to the term deposit that the optimization model could achieve given the constraints and data provided. a higher value indicates a better outcome in terms of the effectiveness of the marketing campaign.
```

```python
# @title .

# Initialize counters for each category
targeted_counts = {'employed': 0, 'student': 0, 'unemployed': 0, 'retired': 0}
contacted_counts = {'employed': 0, 'student': 0, 'unemployed': 0, 'retired': 0}

# Create a dictionary to map client IDs to their corresponding category
client_category_map = {}
for i, client_id in enumerate(client_ids):
    if i < len(category):
        client_category_map[client_id] = category[i]

# Update counts based on optimization results
for v in prob.variables():
    index = int(v.name.split("_")[-1])  # Extract the index regardless of the ID gaps
    client_id = client_ids[index]
    if v.name.startswith("Targeted") and v.varValue == 1:
        category_index = client_category_map.get(client_id, None)
        if category_index:
            targeted_counts[category_index] += 1
    elif v.name.startswith("Contacted") and v.varValue == 1:
        category_index = client_category_map.get(client_id, None)
        if category_index:
            contacted_counts[category_index] += 1

# Print the counts for each category
print("Targeted people by category:")
for category, count in targeted_counts.items():
    print(category, ":", count)

print("\nContacted people by category:")
for category, count in contacted_counts.items():
    print(category, ":", count)
```

# 6 - Scenario 2
In this scenario, the decisionmaker decided to diversify the marketing channels through which he is targeting clients.  Thus, he no longer relies on phone calls solely and decides to explore the possibility of including other means: TV marketing, Print marketing and Email marketing.

**Objective:** Maximize the number of term deposit subscriptions

**Decisions to be made:**  The decision-maker must Decide on the appropriate channels to contact clients


## 6.1 - Mathematical model

#### **Objective Function:**
\begin{align*}
\text{Maximize: } &  \sum_{i=1}^{n} \sum_{c=1}^{4} X_{ic} \cdot P_{ic}
\end{align*}

#### **Constraints:**

**Budget constraint for TV, Print, and Email marketing:**
For each client $i$ and channel $c$, the total cost of targeting through TV, Print, and Email marketing should not exceed the respective budgets.
$$
\sum_{i=1}^{n} (C_c \cdot X_{ic}) \leq B_c \text{ for all } c \in \{1,2,3\}
$$

**Budget constraint for phone calls:**
For each client $i$, the total cost of phone calls should not exceed the phone call budget.
$$
\sum_{i=1}^{n} (D_i \cdot PC \cdot X_{i4}) \leq B_4
$$

**Channels constraint:**
For each channel $c$, the total number of customers targeted should not exceed the maximum allowed for that channel.
$$
\sum_{i=1}^{n} X_{ic} \leq \text{MaxCust}_c, \text{ for all } c \in \{1,2,3,4\}
$$

**Binary variable constraint:**
Each decision variable $X_{ic}$ should be binary, taking values either 0 or 1.
$$
X_{ic} \in \{0, 1\}, \text{ for all } i, c
$$



### 6.1.1 - Preliminary operations

```python
!pip install pulp
import pandas as pd
import pulp
```

### 6.1.2 - Sets


```python
df=pd.read_csv("/content/MDMM - Channels.csv")
clients = df['client_id'].tolist()
channels = ['TV Marketing', 'Print Marketing', 'Email Marketing', 'Phone Calls']
```

### 6.1.3 - Parameters
\begin{align*}
& P_{ic} \in [0, 1]: \text{ expected yield for targeting client } i \text{ through channel } c \\
& B_c: \text{ Budget for channel } c \\
& C_c: \text{ Cost of targeting through channel } c, \text{ for } c \in \{1, 2, 3\} \\
& D_i: \text{ Duration of call for individual } i, \text{ for } i \in \{1, \ldots, n\} \\
& PC: \text{ Phone Call Cost: fixed 0.16 per minute} \\
& \text{MaxCust}_c: \text{ Maximum customers for each channel } c, \text{ for } c \in \{1, 2, 3, 4\}
\end{align*}


```python
# Parameters
channel_costs = {
    'TV Marketing': 2000,
    'Print Marketing': 4000,
    'Email Marketing': 3000,
    'Phone Calls': 0.16  # Cost per minute
}

channel_budgets = {
    'TV Marketing': 10000000,
    'Print Marketing': 2000000,
    'Email Marketing': 3000000,
    'Phone Calls': 5000000
}

max_customers = {
    'TV Marketing': 3000,
    'Print Marketing': 2000,
    'Email Marketing': 1000,
    'Phone Calls': 1000
}

phone_call_cost_per_minute = 0.16
Di = df['duration'].tolist()
```

### 6.1.4 - Variables

**Decision Variables:**
\begin{align*}
& X_{ic} \in \{0, 1\}: \text{ 1 if client } i \text{ is targeted through channel } c, 0 \text{ otherwise.} \\
& \text{Where } i \in \{1, \ldots, n\}: \text{ n is the number of clients in the dataset} \\
& \text{and } c \in \{1, 2, 3, 4\}: \\
& \quad 1: \text{ TV marketing channel} \\
& \quad 2: \text{ Print Marketing} \\
& \quad 3: \text{ Email Marketing} \\
& \quad 4: \text{ Phone Calls}
\end{align*}


```python
# Define the problem
prob = pulp.LpProblem("Maximize_Term_Deposit_Subscriptions", pulp.LpMaximize)

# Decision variables
X = pulp.LpVariable.dicts("X", [(i, c) for i in clients for c in channels], cat='Binary')
```

### 6.1.5 - Constraints and Objective(s)


```python
# Objective function
prob += pulp.lpSum(X[(i, c)] * float(df.loc[df['client_id'] == i, c].iloc[0].replace(',', '.')) for i in clients for c in channels[1:])

# Constraints
for c in channels:
    prob += pulp.lpSum(channel_costs[c] * X[(i, c)] for i in clients) <= channel_budgets[c], f"Budget_constraint_{c}"

for i in clients:
    duration = str(df.loc[df['client_id'] == i, 'duration'].iloc[0]).replace(',', '.')  # Duration for client i
    prob += pulp.lpSum(float(duration) * phone_call_cost_per_minute * X[(i, 'Phone Calls')]) <= channel_budgets['Phone Calls'], f"Phone_call_budget_constraint_{i}"


for c in channels:
    for i in clients:
        prob += pulp.lpSum(X[(i, c)]) <= max_customers[c], f"Max_customers_constraint_{i}_{c}" #add lpsum

for i in clients:
        prob += pulp.lpSum(X[(i, 'Phone Calls')]) ==0, f"phone_constraint_{i}" #add lpsum
# Additional constraint to limit the number of clients targeted through phone calls
prob += pulp.lpSum(X[(i, 'Phone Calls')] for i in clients) <= max_customers['Phone Calls'], "Max_clients_phone_calls_constraint"
```

## 6.2 - Model solution


```python
# Solve the problem
prob.solve()

# Print the status of the solution
print("Status:", pulp.LpStatus[prob.status])

# Print the optimal value of the objective function
print("Optimal value of the objective function:", pulp.value(prob.objective))

# Initialize a dictionary to store channels for each client
client_channels = {}

# Collect channels for each client
for i in clients:
    client_channels[i] = []

for i in clients:
    for c in channels:
        if pulp.value(X[(i, c)]) == 1:
            client_channels[i].append(c)

# Print the channels for each client
for i, channels_list in client_channels.items():
    if channels_list:  # Check if channels_list is not empty
        channels_str = ', '.join(map(str, channels_list))
        print(f"Contact client {i} through channel(s): {channels_str}")
```

```python
# Calculate total number of clients to target per channel
total_clients_per_channel = {c: sum(pulp.value(X[(i, c)]) for i in clients) for c in channels}

# Print the total number of clients to target per channel
for c in channels:
    print(f"Total number of clients to target through {c}: {total_clients_per_channel[c]}")

```

# 7 - Scenario 3

In this scenario, clients who have never been targeted before are prioritized, ensuring they are contacted if feasible within the campaign constraints. Additionally, to avoid annoyance and manage contact frequency, targeted clients should only have been previously contacted at most 2 times.


**Objective:** Maximize the number of term deposits subscriptions.

**Decisions to be made:** The decision-maker must decide on the best period to contact each client.

## 7.1 - Mathematical model

#### **Objective Function:**
\begin{align*}
\text{Maximize: } &  \sum_{i=1}^{n}\sum_{j=1}^{3} X_{ij} \cdot P_{ij}
\end{align*}


 #### **Constraints**

**Budget constraint:** Ensure that the total cost of contacts does not exceed the allocated budget for each period.
$$
\sum_{i=1}^{n} (D_{i} \cdot PC \cdot X_{ij}) \leq B_{j} , \text{ for all }  j \in \{1, 2, 3\}
$$


**Targeted individuals constraint:** The total number of people targeted in each period should not exceed the maximum threshold to manage campaign costs and resources.

$$
\sum_{i=1}^{n} X_{ij} \leq Max_{c}, \quad \text{for all } j \in \{1, 2, 3\}
$$

**Priority targeting constraint:** At least 30% of the clients who have never been contacted should be targeted during each period to prioritize untapped client segments.

$$
\sum_{i=1}^{n} X_{ij} \geq 0.30 \times T_{nc}\quad  ,  \text{ if pdays}[i] = -1 , \quad \text{for all } j \in \{1, 2, 3\}
$$

**Frequency limit constraint:** Limit the frequency of contacts for each client to a maximum of 2 times to avoid annoyance.

$$
\sum_{j=1}^{3} X_{ij} \times \text{previous}[i] \leq 2, \quad \text{for all } i \in \text{clients}, \text{if pdays}[i] \neq -1
$$

**Unique period constraint:** Ensure each client is contacted in at most one period
$$
\sum_{j=1}^{3} X_{ij} \leq 1, \quad \text{for all } i \in \text{clients}
$$

**Binary variable constraint:** Each decision variable $X_{ij}$ should be binary, taking values either 0 or 1.
$$
X_{ij} \in \{0, 1\}, \text{ for all } i, j
$$

### 7.1.1 - Preliminary operations


```python
import pandas as pd
import pulp
from pulp import *
```

### 7.1.2 - Sets

```python
# Read the dataset from Excel file
df=pd.read_csv("/content/MDMM - Periods (2).csv")

# Extract relevant columns from the DataFrame
client_ids = df['client_id'].tolist()
clients = range(len(df))
period = ['1', '2', '3']
periods = range(1, 4)
```

### 7.1.3 - Parameters
\begin{align*}
& P_{ij} \in [0, 1]: \text{ expected yield for targeting client } i \text{ in period } j \\
& B_j: \text{ Budget allocated for period } j \\
& previous_i: \text{ The number of times a client has ever been contacted }\\
& pdays_i: \text{ The days that have passed since the last contact of client} i\\
& D_i: \text{ Duration of call for individual } i, \text{ for } i \in \{1, \ldots, n\} \\
& PC: \text{ Phone Call Cost: fixed 0.16 per minute} \\
& Max_{c}: \text{ Maximum number of clients to be contacted in each period} \\
& T_{nc}: \text{ Total number of clients that have never been contacted before} \\
\end{align*}


```python
for col in ['1', '2', '3']:
    df[col] = df[col].str.replace(',', '.')
probabilities = {row['client_id']: {1: float(row['1']), 2: float(row['2']), 3: float(row['3'])} for index, row in df.iterrows()}
budgets = {1: 200000,
           2: 100000,
           3: 80000}
duration = df['duration'].tolist()
pdays = df['pdays'].tolist()
previous = df['previous'].tolist()
PC = 0.16
Maxc= 3000
Tnc=7070
```

### 7.1.4 - Variables

**Decision Variables:**
\begin{align*}
& X_{ij} \in \{0, 1\}: \text{ 1 if client } i \text{ is targeted in period } j, 0 \text{ otherwise.} \\
& \text{Where } i \in \{1, \ldots, n\}: \text{ n is the number of clients in the dataset} \\
& \text{and } j \in \{1, 2, 3\}: \\
& \quad 1: \text{  from September to December} \\
& \quad 2: \text{ from January to April}  \\
& \quad 3: \text{ from May to August}\\
\end{align*}

```python
# Decision variables
X = LpVariable.dicts("Target_Client", (clients, periods), 0, 1, LpBinary)
```

### 7.1.5 - Constraints and Objective(s)

```python
# Define the optimization problem
prob = LpProblem("Maximize_Term_Deposit_Subscriptions", LpMaximize)

# Objective function
prob += lpSum(X[i][j] * probabilities[client_ids[i]][j] for i in clients for j in periods)

# Constraints

# Ensure each client is contacted in only one period
for i in clients:
    prob += lpSum(X[i][j] for j in periods) <= 1, f"Client_Unique_Period_{i}"

# Ensure that the max number of customers contacted in each period is 3000
for j in periods:
    prob += lpSum(X[i][j] for i in clients) <= Maxc, f"Max_Clients_Constraint_Period_{j}"

# Ensure that at least 30% of the clients who were never contacted get contacted in each period
for j in periods:
    never_contacted_targeted_period = lpSum(X[i][j] for i in clients if pdays[i] == -1)
    prob += never_contacted_targeted_period >= 0.30 * Tnc, f"At_Least_30Percent_Contacted_Never_Constraint_Period_{j}"

# Ensure that the maximum number contacts for each client is 2 times
for i in clients:
    if pdays[i] != -1:
        prob += lpSum(X[i][j] * previous[clients[i]] for j in periods) <= 1, f"Max_pdays_Constraint_Client_{i}"

# Budget constraint
for j in periods:
    total_cost_period = lpSum(X[i][j] * duration[clients[i]] * 0.16 for i in clients)
    prob += total_cost_period <= budgets[j], f"Budget_Constraint_Period_{j}"
```

## 7.2 - Model solution


```python
# Solve the optimization problem
prob.solve()

# Display the final output
print("Status:", LpStatus[prob.status])
print("Final Output:")
# Print the objective function value (the total expected yield)
print("Objective Function Value:", value(prob.objective))
for j in periods:
    clients_targeted = [client_ids[i] for i in clients if value(X[i][j]) == 1]
    print(f"Period {j}: Target {len(clients_targeted)} clients - {clients_targeted}")
```

# 8 - Scenario 4

**Objective:** Maximize the number of term deposits subscriptions.

**Decisions to be made:** The decision-maker must decide the best type of clients to target in his next campaign.

## 8.1 - Mathematical model

**Objective Function:**
\begin{align*}
\text{Maximize: } & \sum_{j}\sum_{k} T_{jk} \cdot P_{jk}
\end{align*}

 #### **Constraints:**
**Budget constraint:** The total costs of calls of the targeted individuals should not exceed the fixed budget TB:
$$
\sum_{j=1}^{3} \sum_{k=1}^{2}  (D_{jk}\cdot PC) \cdot T_{jk} \leq TB
$$

**Default history:** Individuals must be targeted in a way that the defaulted balance threshold for category jk is not exceeded.

$$
T_{jk} \cdot DF_{jk} \leq DT_j, \forall j, k
$$

**Duration constraint:** Individuals must be targeted in a way that the total duration of calls for category jk is not exceeded.

$$
T_{jk} \cdot D_{jk} \leq DM_j, \forall j, k
$$

**Targeted individuals constraint:**  The total number of people targeted in the subcategory jk should exceed the minimum thershold of targets in that subcategory
$$
\sum_{k=1}^2 T_{jk} \leq DM_j, \forall j
$$

**Positivity constraint:** Each decision variable $T_{jk}$ should be positive
$$
T_{jk} \geq 0, \quad \forall j \in \{1,2,3\}, \text{ and } k \in \{1,2\}
$$

### 8.1.1 - Preliminary operations


```python
!pip install pulp
import pandas as pd
import pulp
from pulp import *
```

### 8.1.2 - Sets


```python
# Read the dataset
data = pd.read_excel('Probabilities MDMM.xlsx')
```

### 8.1.3 - Parameters

$P_{jk}$: Probability of subscription of subcategory of type $jk$, where $P_{jk} \in [0,1]$. \\
$D_{jk}$: Duration of call in the subcategory of type $jk$. \\
$DF_{jk}$: Default in subcategory $jk$. \\
$PC$: Phone Call Cost (per minute) = 0.16 eur/minute. \\
$DT_{j}$: Default Threshold per category $j$. \\
$TB$: Total budget for the campaign = 5000000. \\
$DM_{j}$: Maximum duration for category $j$. \\
$MinCat_{j}$: Minimum individuals to be chosen from category $j$. \\

```python
# Parameters
Pjk = data['Probabilities'].tolist()
Djk = data['duration'].tolist()
DFjk = data['default'].tolist()
PC = 0.16  # per minute cost
TB = 25000000  # total budget

Maximum_default = {
    'divorced': 30000,
    'married': 2000000,
    'single': 900000,
}

Maximum_duration = {
    'divorced': 50000,
    'married': 400000,
    'single': 150000,
}

Minimum_individuals = {
    'divorced': 200,
    'married': 0,
    'single': 500,
}
```

### 8.1.4 - Variables

**Decision Variable:**
\begin{align*}
& T_{jk}:\text{The number of customers for each subcategory combination jk} \\
& \text{Where } j \in \{1, 2, 3\}: \\
& \quad 1: \text{divorced} \\
& \quad 2: \text{married} \\
& \quad 3: \text{single} \\
& \text{and } k \in \{1, 2\}: \\
& \quad 1: \text{1 or has college level education} \\
& \quad 2: \text{0 or does not have college level education} \\
\end{align*}

```python
# Decision variable
Tjk = LpVariable.dicts("T", [(j, k) for j in range(1, 4) for k in range(1, 3)], lowBound=0, cat='Integer')
```

### 8.1.5 - Constraints and Objectives:


```python
# Define the optimization problem
prob = LpProblem("Maximize_Subscriptions", LpMaximize)

# Objective function
prob += lpSum([Tjk[j, k] * Pjk[j-1+k] for j in range(1, 4) for k in range(1, 3)])

# Constraints

# Budget constraint
prob += lpSum([Djk[j-1+k] * PC * Tjk[j, k] for j in range(1, 4) for k in range(1, 3)]) <= TB

# Default history constraint
for j in range(1, 4):
    for k in range(1, 3):
        prob += Tjk[j, k] * DFjk[j-1+k] <= Maximum_default[list(Maximum_default.keys())[j-1]]

# Duration per category constraint
for j in range(1, 4):
    for k in range(1, 3):
        prob += Tjk[j, k] * Djk[j-1+k] <= Maximum_duration[list(Maximum_duration.keys())[j-1]]

# Duration per category constraint
for j in range(1, 4):
    for k in range(1, 3):
        prob += Tjk[j, k] * Djk[j-1+k] <= Maximum_duration[list(Maximum_duration.keys())[j-1]]

# Minimum individuals constraint
for j in range(1, 4):
    prob += lpSum([Tjk[j, k] for k in range(1, 3)]) >= Minimum_individuals[list(Minimum_individuals.keys())[j-1]]

# Minimum individuals constraint
prob += lpSum([Tjk[2, k] for k in range(1, 3)]) <= 0
```

```python
# Solve the problem
prob.solve()

# Print the results
print("Status:", LpStatus[prob.status])
print("Objective value (Max Sum):", value(prob.objective))

# Print the optimal values of decision variables
for v in prob.variables():
    print(v.name, "=", v.varValue)
```
